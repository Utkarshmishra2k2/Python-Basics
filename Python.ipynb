#Prime Number

Method 01

```python
minimum = int(input("Enter the min : "))
maximum = int(input("Enter the max : "))
print("Prime Number's are")
for n in range(minimum,maximum + 1):
    if n > 1:
        for i in range(2,n):
            if (n % i) == 0:
                break
        else:
            print(n)
```

#Fabonacci Series

```python
def fab(n):
    if n == 0 :
        return 0
    elif n==1:
        return 1
    elif n==2:
        return 1
    else:
        return fab(n-1)+fab(n-2)
for i in range(15):
    print(fab(i), end = ', ')
```

#Factorial

Finding Factorial of Series

```python
def fact(num):
    if num<0:
        return "number invalid"
    elif num == 1:
        return 1
    elif num == 0:
        return 0
    else:
        return num * fact(num-1)
for i in range(5):
    print(fact(i))
```

#Amstrong Number

Finding Amstrong Number

```python
num = int(input("Enter a number : "))
temp = num
add = 0
while temp > 0 :
    rem = temp%10
    add = add + rem**3
    temp = temp//10

if num == add:
    print('Its Amstrong Number')
else:
    print('Its Not Amstrong')
```

#Fizz Buzz Method

```python
num = int(input('Enter a number : '))
for i in range(1,num+1):
    if i%3 == 0:
        print('Fizz')
    elif i%5 == 0:
        print('Buzz')
    elif i%5 == 0 and i%3 == 0:
        print('FizzBuzz')
    else:
        print(i)
```

```python
def fizz(num):
    dic = {3:'fizz',5:'buzz'}
    for i in range(1,num+1):
        result = ''
        for k,v in dic.items():
            if i%k==0:
                result = result+v
        if not result :
            result = i
        print(result)
fizz(15)
```

#Grouping

You are given a string "S". Suppose a character "c"' occurs consecutively "X"times in the string. Replace these consecutive occurrences of the character ''c" with (X,c) in S should be a number

```python
from itertools import groupby
for k, c in groupby(input("Enter a Number : ")):
    print((len(list(c)), int(k)), end=' ')
```
